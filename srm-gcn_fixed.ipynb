```python
import numpy as np
import pandas as pd
import os 
import os.path as osp
import random
import torch
import torch.nn.functional as F
from torch_geometric.datasets import Planetoid
import torch_geometric.transforms as T
from torch_geometric.nn import GCNConv, ChebConv  # noqa
from pathlib import Path
from torch_geometric.data import Data

dir = str(Path.cwd())
encodings = pd.read_csv(dir+"/encodings(catx-orix)2.csv")
print(encodings.iloc[0][0])
#average encoding score of the graphs below
avg=.05276894119
adjs = np.load(dir+"/adjs.npy", allow_pickle=True).item()   
                      
```

```python
beg=[]
end=[]
data_list=[]
row=0
col=0
adj=0

x = torch.Tensor(np.zeros((1000,100))).to('cpu')
x=x+5
print(x)

#
# labels graph 1 if above average, 0 if below average
y = []
for ei, _ in enumerate(adjs):
    if encodings.iloc[ei][1] > .05276894119:
        y.append(0)
    else:
        y.append(1)
y = np.array(y)
y = torch.Tensor(y).to('cpu')

# iterates through each graph adjacency matrix, converts graph into 
# COO format as described in pytorch geom documentation. this is edge_index
# x contains the feature matrix for each node in the graph
# y is the graph label (0 or 1)
for adj_i in adjs:
    row=0
    adj_m = adjs[adj_i]
    nn, mm = adj_m.shape
    full_adj = np.vstack([np.hstack([np.zeros((nn, nn)), adj_m]), np.hstack([adj_m.T, np.zeros((mm, mm))])]) #reshape 100x100
    for line in full_adj:
        #print(line)
        col=0
        for num in line:
            if num == 1:
                beg.append(row)
                end.append(col)
            col+=1
        row+=1
    edge_index = torch.tensor([beg, end], dtype=torch.long)
    #print(edge_index)
    data = Data(edge_index = edge_index, x=x, y=y)
    #print(data.y)
    data.num_nodes=100
    data_list.append(data)
    beg=[]
    end=[]
    adj+=1
#print(data_list)
```
